```python
# step 1: Notes data
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.\n"
    "Topic 2: Lists are ordered and mutable.\n"
    "Topic 3: Dictionaries store key-value pairs.\n"
    "Topic 4: Loops automate repetitive tasks.\n"
    "Topic 5: Exception handling prevents crashes.\n"
]

#step 2: Write to file
with open("python_notes.txt", "w", encoding="utf-8") as file:
  file.writelines(notes)

print("File written successfully.")

#step 3: Append more lines
with open("python_notes.txt", "a", encoding="utf-8") as file:
  file.write("Topic 6: Functions help reuse code.\n")
  file.write("Topic 7: Modules help organize code.\n")
print("Lines append.")

#step 4: Read file
with open("python_notes.txt", "r", encoding="utf-8") as file:
  lines = file.readlines()

#step 5: Print numbered lines
print("\n--- File content ---")
for i, line in enumerate(lines, start=1):
  print(f"{i}. {line.strip()}")

#step 6: Count lines
print("\nTotal lines:", len(lines))

#step 7: Keyword search
keyword = input("\nEnter a keyword to search for: ").lower()

found = False
for line in lines:
  if keyword in line.lower():
    print(line.strip())
    found = True

if not found:
  print("No matching lines found.")
```

```python
import requests

#----------------- Step 1 -----------------
print("\nStep 1: Fetch Products")

url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)
data = response.json()

products = data["products"]

print("ID | Title | Category | Price | Rating")
print("-" * 60)

for p in products:
  print(f"{p['id']} | {p['title']} | {p['category']} | ${p['price']} | {p['rating']} ")

#---------------- Step 3 ---------------
print("\nSTEP 3: Laptops Category")

url2 = "https://dummyjson.com/products/category/laptops"
response2 = requests.get(url2)
data2 = response2.json()

laptops = data2["products"]

print("Title | Price")
print("-" * 40)

for item in laptops:
  print(f"{item['title']} | ${item['price']}")

#--------------- Step 4 ---------------
print("\nSTEP 4: POST Request")

url3 = "https://dummyjson.com/products/add"

new_product = {
  "title": "My CUstom Product",
  "price": 999,
  "category": "electronics",
  "description": "A product I created via API"
}

response = requests.post(url3, json=new_product)

print("Response:")
print(response.json())

```

```python
import requests

# Part A
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"
    except TypeError:
        return "Invalid input types"

# Part B
def read_file_safe(filename):
    try:
        with open(filename, "r") as file:
            return file.read()
    except FileNotFoundError:
        return f"File '{filename}' not found"
    finally:
        print("File operation done")

# Part C
def fetch_data(product_id):
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)
        return response
    except Exception as e:
        return f"Error: {e}"

# Part D (NO INPUT LOOP)
def product_lookup(product_id):
    response = fetch_data(product_id)

    if isinstance(response, str):
        print(response)
        return

    if response.status_code == 200:
        data = response.json()
        print("Product:", data["title"])
        print("Price:", data["price"])
    else:
        print("Product not found")

#  RUN TESTS
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))

print(read_file_safe("python_notes.txt"))

#  Direct call (NO INPUT)
product_lookup(5)
```

```python
import requests
from datetime import datetime

#----------- LOG FUNCTION ----------
def log_error(message):
  with open("error_log.txt", "a") as file:
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    file.write(f"[{timestamp}] {message}\n")

#--Connection error--
def fetch_products():
  try:
    url ="https://this-does-not-exist-xyz.com/api"
    requests.get(url, Timeout=5)
  except requests.exceptions.connectionerror:
    log_error("ERROR in fetch_products: ConnectonError - No connection could be made")

#-- HTTP ERROR --
def lookup_product(product_id):
  try:
    url = f"https://dummyjson.com/products/{product_id}"
    response = requests.get(url)

    if response.status_code == 404:
      log_error(f"ERROR in lookup_product: HTTPError -404 not found for product Id{product_id}")
  except Exception as e:
      log_error(f"ERROR in lookup_product:{e}")

#-- DISPLAY LOGS --
def display_logs():
    try:
      with open("error_log.txt", "r") as file:
        print("\n-- ERROR LOG CONTENT --")
        print(file.read())
    except FileNotFoundError:
          print("No logs found.")

#-- RUN PROGRAM --
fetch_data("http://this-host-does-not-exist-xyz.com/api")                 # triggers connection error
lookup_product(999)          # triggers 404 error
display_logs()               # show log file
```
